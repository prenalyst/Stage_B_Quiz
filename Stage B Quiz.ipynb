# Separating the columns into similar groups to facilitate intuition

```python
col_temp = ["T1","T2","T3","T4","T5","T6","T7","T8","T9"]

col_hum = ["RH_1","RH_2","RH_3","RH_4","RH_5","RH_6","RH_7","RH_8","RH_9"]


col_weather = ["T_out", "Tdewpoint","RH_out","Press_mm_hg",
                "Windspeed","Visibility"] 

col_random = ['rv1', 'rv2']

col_target = 'Appliances'
```

## Normalizing the features in the dataset using MinMaxScaler

```python
scaler = MinMaxScaler()

normed_data = pd.DataFrame(scaler.fit_transform(data), columns=data.columns)
normed_data
```

## Separate dependent and independent variables

```python
feature_vars = normed_data.drop(columns=['rv1', 'rv2', 'Appliances'])
target_var = normed_data[col_target]
```

## Split dataset into train and test set

```python
x_train, x_test, y_train, y_test = train_test_split(feature_vars, target_var, train_size=0.7, random_state=42)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)

# predicting the output of the training set
# using the Linear Regression model
y_train_predict = lr.predict(x_train)

# predicting the output of the test set
# using the Linear Regression model
y_test_predict = lr.predict(x_test)

test_mse = mean_squared_error(y_test, y_test_predict)

# calculate root mean squared value

test_rmse = np.sqrt(test_mse)

print(test_rmse)
```

# Question 2

Train a lasso regression model with an alpha value of 0.001 and obtain the new feature weights with it. How many of the features have non-zero feature weights?

### Solution:

```python
# instantiating Lasso
lasso = Lasso(alpha=0.001)

# fitting the training data
lasso.fit(x_train, y_train)


# create a pandas series with the coefficients as the values
# using the column names of x_train as the index
weights = pd.Series(lasso.coef_, x_train.columns)

# use boolean indexing to return values
# that are NOT equal to zero
weights = weights[weights != 0]

# print the length of the returned value
print('Features with non-zero weights:', len(weights))
```
